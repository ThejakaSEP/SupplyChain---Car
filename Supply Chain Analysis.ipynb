```python
# Convert columns to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
```

```python
df['Delivery Time'] = (df['ShipDate'] - df['OrderDate'])
df['Delivery Time']
```

```python
# We can see that there are negative values for the difference
# This suggests that either the data is incorrect or has been switched between the columns.
# For this instance we will consider as the data got switched and let's correct them


for i in df.index:
    if df['Delivery Time'][i].days < 0:
        x = df['ShipDate'][i]
        df['ShipDate'][i] = df['OrderDate'][i]
        df['OrderDate'][i] = x
        df['Delivery Time'][i] = (df['ShipDate'][i] - df['OrderDate'][i])
        print(df['Delivery Time'][i])
       
       
```

```python
#Convert Days to Integer for easy plotting
df['Delivery Time'] = pd.to_numeric(df['Delivery Time'].dt.days, downcast='integer')
```

```python
fig_dims = (40, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.histplot(data=df, x="Delivery Time",ax=ax,bins = 100)

```
